```python
from fussballgott import load, league, plot
```

```python
file = 'demo_league.CSV'
teams, schedule, table, missing_games = load.load(file, style = 'league')
```

```python
sim = league.simulate(teams, schedule, table, missing_games, n_sim = 1e5)
```

```python
plot.league(sim)
```

```python
file_sched = 'demo_league_sep_schedule.CSV'
file_tab = 'demo_league_sep_table.CSV'
teams, schedule, table, missing_games = load.load([file_sched, file_tab], style = 'league_sep')
```

```python
sim = league.simulate(teams, schedule, table, missing_games, n_sim = 1e5)
```

```python
plot.league(sim)
```
